# Exploring Ebay Car Sales Data

we'll work with a dataset of used cars from eBay Kleinanzeigen, a classifieds section of the German eBay website.


- dateCrawled - When this ad was first crawled. All field-values are taken from this date.
- name - Name of the car.
- seller - Whether the seller is private or a dealer.
- offerType - The type of listing
- price - The price on the ad to sell the car.
- abtest - Whether the listing is included in an A/B test.
- vehicleType - The vehicle Type.
- yearOfRegistration - The year in which the car was first registered.
- gearbox - The transmission type.
- powerPS - The power of the car in PS.
- model - The car model name.
- kilometer - How many kilometers the car has driven.
- monthOfRegistration - The month in which the car was first registered.
- fuelType - What type of fuel the car uses.
- brand - The brand of the car.
- notRepairedDamage - If the car has a damage which is not yet repaired.
- dateCreated - The date on which the eBay listing was created.
- nrOfPictures - The number of pictures in the ad.
- postalCode - The postal code for the location of the vehicle.
- lastSeenOnline - When the crawler saw this ad last online.

Following Changes are made in column names:
- yearOfRegistration to registration_year
- monthOfRegistration to registration_month
- notRepairedDamage to unrepaired_damage
- dateCreated to ad_created
- All Column names are converted from camelcase to snakecase

```python
autos.columns = ['date_crawled', 'name', 'seller', 'offer_type', 'price', 'ab_test',
       'vehicle_type', 'registration_year', 'gear_box', 'power_ps', 'model',
       'odometer', 'registration_month', 'fuel_type', 'brand',
       'unrepaired_damage', 'ad_created', 'num_photos', 'postal_code',
       'last_seen']
```

# Initial Exploration and Cleaning

Descriptive Statistics

```python
autos.describe(include="all")
```

Seller , Offer Type are candidates for deletion as they mostly contain one value.
Price , Odometer are numeric values stored as text.



```python
autos["num_photos"].value_counts()
```

```python
#num of photos also only has 0 value . Let's delete all above 3 columns

autos.drop(["seller","offer_type","num_photos"],axis=1,inplace=True)
```

```python

autos["price"] = (autos["price"].str.replace("$","").str.replace(",","").astype(int))
```

```python
# removing non-numeric text from Price , Odometer

autos["odometer"] = (autos["odometer"].str.replace("km","").str.replace(",","").astype(int))

```

```python
autos.rename({"odometer":"odometer_km"},axis=1,inplace=True)
```

# Dealing with Incorrect Registration Year Data

Because a car can't be first registered after the listing was seen, any vehicle with a registration year above 2016 is definitely inaccurate. Determining the earliest valid year is more difficult. Realistically, it could be somewhere in the first few decades of the 1900s.

```python
autos = autos[autos['registration_year'].between(1900,2016)]
```

```python
autos['registration_year'].value_counts(normalize =True).head(10)
```

Majority of Cars were registered in last 10 years.

# Exploring Price by Brand

```python
autos["brand"].unique().shape
```

```python
autos["brand"].value_counts().head(20)
```

```python
top_20_brands = autos["brand"].value_counts().head(20).index
```

```python
autos["brand"].value_counts(normalize = True).head(20)
```

We will explore top 20 brands . We can see that top 10 includes majority of the cars.

```python
brand_mean_price = {}

for row in top_20_brands:
    brand_mean_price[row] = np.mean(autos[autos["brand"] == row]["price"])
    
```

```python
brand_mean_price
```

sonstige_autos is the most expensive brand with mean price 12338 & renault is cheapest with 2474.

```python
brand_mean_price = {}
brand_mean_mileage = {}
for row in top_20_brands:
    brand_mean_price[row] = np.mean(autos[autos["brand"] == row]["price"])
    brand_mean_mileage[row] = np.mean(autos[autos["brand"] == row]["odometer_km"])
    
```

```python
brand_mean_price = pd.Series(brand_mean_price)
brand_mean_mileage = pd.Series(brand_mean_mileage)
```

```python
df = pd.DataFrame(brand_mean_price, columns=['mean_price'])
```

```python
df["mean_mileage"] = brand_mean_mileage
```

```python
df
```

The range of car mileages does not vary as much as the prices do by brand, instead all falling within 10% for the top brands. There is a slight trend to the more expensive vehicles having higher mileage, with the less expensive vehicles having lower mileage.
